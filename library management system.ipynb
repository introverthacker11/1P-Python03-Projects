```python
while True :

    pass_ = 'Free Palestine'
    password = input('Enter password: ')

    if pass_ == password :

        print('\nAccess Granted!\n')

        import time

        currenttime = time.localtime()
        time = time.strftime('%d-%m-%Y,%H:%M:%S', currenttime)

        books = ['Quran-e-Majeed','Hadees-e-Nabvi(S.A.W,P.B.U.H)','To Kill a Mockingbird by Harper Lee','The Elements of Style by William Strunk Jr. and E.B. White','The Origin of Species by Charles Darwin','The Wealth of Nations by Adam Smith','1984 by George Orwell','The Art of War by Sun Tzu','The Diary of Anne Frank by Anne Frank', "The Hitchhiker's Guide to the Galaxy by Douglas Adams",'The Complete Works of William Shakespeare']
        print('=' * 120)
        print('\nTime(Karachi): ',time)
        print('\nWelcome to Rayyan Ahmed\'s \"Python3 Based Library Management System\".')
        print('\U0001F970' * 3)
        print(f'\nBooks Availabled:\n\n{books}')

        books_purchased = []

        while True :

            choice = input('\nWant to: a)Buy Book...b)Return Book...c)Exit Program.\nEnter your choice(a/b/c): ')

            if choice == 'a' :

                print('\nBooks Availabled: ', books)
                buy_book = input('\nWhich book you want to purchased: ')

                if buy_book in books :
                    books.remove(buy_book)
                    books_purchased.append(buy_book)
                    print('\nStatus: Purchase Done!' + '\U0001F60d')
                    print(f'Details:\nBook Purchased: {buy_book}...Purchase Time(day-month-year): {time}')

                else :
                    print('\nStatus: Purchase Failed')
                    print(f'The book \"{buy_book}\" is not availabled right now!')
                    print('\U0001F625')

            elif choice == 'b' :

                print(f'\nBooks to be returned: {books_purchased}')
                return_book = input('\nWhich book you want to return: ')

                if return_book in books_purchased :
                    books_purchased.remove(return_book)
                    books.append(return_book)
                    print('\nStatus: Return Done!')
                    print(f'Details:\nBook Returned: {return_book}...Time when book is returned: {time}')
                    print('Thanks Alot! We appreciate it.')
                    print('\U0001F917')

                else :
                    print(f'\nThe book \"{return_book}\" is not in your purchase list.')
                    print('\U0001F62A')

            elif choice == 'c' :
                print('\nThanks for visiting our library.\nTake Care, Allah Hafiz.')
                print('\U0001F600')
                break

            else :
                print('\nEnter valid input!\n(a/b/c)')

    else :
        print('\nX--Alert!--X')
        print('\nAccess Denied!')
        print('\U0001F620' * 3)
    
    run = input('\nWanna run again (y/n): ')
    if run != 'y' :
        print('\nYou Choose to exit.Good Bye.\n')
        print('\U0001F600')
        break
```
