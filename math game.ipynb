```python
import random

while True :
    
    try :
        
        print('Welcome Here!\n')
        pass_ = 'Free Palestine'
        passw = input('Enter password: ')
        
        if pass_ == passw :
            
            score = 0
            tries = 0
            q = 0
            name = input('Access Granted!\n\nWhat\'s your name: ')
            try_ = int(input(f'{name}! How many tries would you like: '))
            print(f'\nGame has been started.\nGood Luck {name}!')
            
            while try_ != tries :
                
                num1 = random.randint(0,1)
                num2 = random.randint(0,10)
                operator = ['+','-','*']
                oper = random.choice(operator)
                q += 1
                
                if oper == '+' :
                    
                    guess = int(input(f'\nWhat is the sum of {num1} + {num2}: '))
                    result = num1 + num2
                    
                elif oper == '-' :
                    
                    guess = int(input(f'\nWhat is the subtraction of {num1} - {num2}: '))
                    result = num1 - num2
                    
                elif oper == '*' :
                    
                    guess = int(input(f'\nWhat is the product of {num1} * {num2}: '))
                    result = num1 * num2
                    
                else :
                    print('By.')
                    
                if result == guess :
                    
                    score += 1
                    tries += 1
                    print(f'Score: {score}.')
                    
                elif result != guess :
                    
                    score += 0
                    tries += 1
                    print(f'Score: {score}.')
                    
                else :
                    print('Wrong Input!')
                
            perc = (score / tries) * 100
            print('\nGame has ended now!')
            print( ('_' * 30 ) + 'X-X-X' + ('_' * 30 ))
            
            print(f'\nYour Record:\nScore: {score}\nTries: {try_}\nWin Percentage: {perc:.3f}% \n')
            
            if perc > 50 :
                print(f'Congratulations {name}! You passed the game.')
                
            else :
                print(f'Sorry {name}! You couldn\'t pass the game but better luck for next time.')

        else :
            print('Access Denied!')
            
    except :
        print('Enter valid input!')
        
    run = input('Do you wanna test your brain once more!\n(y/n)?: ')
    
    if run != 'y' :
        print('\nYou choosed to exit.\nAllah Hafiz.')
        break
```
