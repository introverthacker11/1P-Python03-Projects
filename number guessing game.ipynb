```python
import random

while True :

    try :

        sec_num = random.randint(0,100)
        print('_' * 120)
        print('\n')
        print(f'\nYou have entered \"Rayyan Ahmed\'s\" Python3 Program.\n')
        print('_' * 120)
        print('\nWelcome to \"Python3\" based number guessing game.')
        name = input('Enter your name: ')
        tries = int(input('\nHow many tries would you like to get: '))
        print(f'\nDear {name}! In this game, you have to guess the secret number which is from 0 to 100 within {tries} tries.\nYour game has started now.')
        print('\nBest of the luck!')
        try_ = 0


        try :

            while try_ != tries :

                guess = int(input('\nGuess the secret number which is from 0 to 100: '))
                try_ += 1

                if -1 < guess < 101 :

                    if guess < sec_num :
                        print(f'\nYour Guess {guess} is less than the secret number!\nTry with a bigger number.You have {tries - try_} tries left.')
                    elif guess > sec_num :
                        print(f'\nYour Guess {guess} is greator than the secret number!\nTry with a bigger number.You have {tries - try_} tries left.')
                    elif guess == sec_num :
                        print(f'\nCongratulations {name}!\nYour guess {guess} is correct.You have won the game!\nYou passed the game in {try_} tries.')
                        
                        break
                    else :
                        print('\nEnter valid input!')
                        break

                else :
                    print('\nGuess must from 1 to 100.')

            if try_ == tries :

                print('\nYour tries are ended now!')
                
                if guess != sec_num :
                    print(f'\nSorry {name}!.\nYou couldn\'t pass the game this time.\nYou can play it again.')
                elif guess == sec_num :
                    print(f'\nCongratulations {name}!\nYou have won the game!\nYou passed the game in {try_} tries.')

            else :
                None

        except :
            print(f'\n{name} please enter a valid input.')
            
            
    except :
        print('\nEnter valid input.')
        
        
    run = input(f'\n{name}!\nWanna play again: ')

    if run != 'y' :
        print(f'\nYou choosed to exit.\nGood Bye {name}.')
        break
```
