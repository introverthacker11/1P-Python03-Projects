```python
while True :
    
    try :
        
        print('\nWelcome to Rayyan Ahmed Python3 Based Parking Management System.\n')
        pass_ = 'Free Palestine'
        passw = input('Enter the password run program: ')
        
        if pass_ == passw :
            
            nob = 0
            noc = 0
            nobk = 0
            noat = 0
            notx = 0
            nosp = 100
            total_fee = 0
            print('Access Granted!\n')
            
            while True :
                
                print('\nWhat you want to park?\na)Auto...b)Bus...c)Car...d)Taxi...e)Bike...\nf)Show Status...g)Delete data...h)Exit')
                choice = input('Enter your choice (a,b,c,d,e): ')
                
                if choice == 'a' :
                    print(f'\nEntry of auto succed! Ticket charged: 20 PKR')
                    noat += 1
                    nosp -= 1
                    total_fee += 20
                    
                elif choice == 'b' :
                    print(f'\nEntry of bus succed! Ticket charged: 50 PKR')
                    nob += 1
                    nosp -= 1
                    total_fee += 50
                    
                elif choice == 'c' :
                    print(f'\nEntry of car succed! Ticket charged: 80 PKR')
                    noc += 1
                    nosp -= 1
                    total_fee += 80
                    
                elif choice == 'd' :
                    print(f'\nEntry of taxi succed! Ticket charged: 30 PKR')
                    notx += 1
                    nosp -= 1
                    total_fee += 30
                    
                elif choice == 'e' :
                    print(f'\nEntry of bike succeed! Ticket charged: 40 PKR')
                    nobk += 1
                    nosp -= 1
                    total_fee += 40
                    
                elif choice == 'f' :
                    print('\nParking Status:\n')
                    print(f'No. of Cars: {noc}\nNo. of Bikes: {nobk}\nNo. of Buses: {nob}\nNo. of Autos: {noat}\nNo. of Taxies: {notx}\nTotal Vehicles Parked: {nobk + noc + notx + noat + nob}\nSpace Availabled in Parking: {nosp}\nTotal Fee Collected: {total_fee}.')
                    
                elif choice == 'g' :
                    nob = 0
                    noc = 0
                    nobk = 0
                    noat = 0
                    notx = 0
                    nosp = 100
                    total_fee = 0
                    print('\nParking Status after deleting Data:\n')
                    print(f'No. of Cars: {noc}\nNo. of Bikes: {nobk}\nNo. of Buses: {nob}\nNo. of Autos: {noat}\nNo. of Taxies: {notx}\nTotal Vehicles Parked: {nobk + noc + notx + noat + nob}\nSpace Availabled in Parking: {nosp}\nTotal Fee Collected: {total_fee}.')

                    
                elif choice == 'h' :
                    print('\nYou choosed to exit,good bye!')
                    break
                    
                else :
                    print('\nEnter valid choice.')
                    
        else :
            print('\nAccess Denied! Try again')
                
    except :
        print('\nInvalid Input!')
        
    run = input('\nWanna run again (y/n)?: ')
    
    if run != 'y' :
        print('\nYou choosed to exit.\nGood Bye.')
```
