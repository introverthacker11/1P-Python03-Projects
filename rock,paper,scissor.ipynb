```python
import random

try :
    
    print('=' * 120)
    print('\nAssalam-o-alaikum!\nWelcome to Rayyan Ahmed\'s \"Python3 Based Rock,Paper,Scissor Game\"')
    name = input('\nEnter your name: ')
    tries = int(input('How many tries would you like to get?: '))
    print(f'\n{name}! Now you are playing against Python3 program.Score more than program to win the game.')
    print('\nNote: Guess except Rock, Paper or Scissor will result in increasing program\'s score by 1.')
    try_ = 0
    y_score = 0 #your score
    p_score = 0 #program score
    a = '\nAttempt won.'
    b = '\nAttempt draw.'
    c = '\nAttempt lost.'
    d = '\nInvalid Guess! Program got 1 score due to your invalid input.'

    while try_ != tries :

        list_ = ['Rock','Paper','Scissor']
        ans = random.choice(list_)
        print(list_)

        guess = input('\n===>>Enter your guess: ')
        try_ += 1
        print(f'\nYour Guess: {guess}.')
        print(f'Program\'s Guess: {ans}.')

        if ans == 'Rock' :

            if guess == 'Rock' :
                print(b)
                y_score += 0
                p_score += 0

            elif guess == 'Scissor' :
                print(c)
                y_score += 0
                p_score += 1

            elif guess == 'Paper' :
                print(c)
                y_score += 0
                p_score += 1
                
            else :
                print(d)
                p_score += 1

        elif ans == 'Scissor' :

            if guess == 'Rock' :
                print(a)
                y_score += 1
                p_score += 0

            elif guess == 'Scissor' :
                print(b)
                y_score += 0
                p_score += 0

            elif guess == 'Paper' :
                print(c)
                y_score += 0
                p_score += 1  
                
            else :
                print(d)
                p_score += 1

        elif ans == 'Paper' :

            if guess == 'Rock' :
                print(a)
                y_score += 1
                p_score += 0

            elif guess == 'Scissor' :
                print(a)
                y_score += 1
                p_score += 0

            elif guess == 'Paper' :
                print(b)
                y_score += 0
                p_score += 0
                
            else :
                print(d)
                p_score += 1

    print(f'\n{tries} tries comleted!')
    print(f'Scorecard:\nYour Score: {y_score}.\nProgram\'s Score: {p_score}.')

    if y_score > p_score :
        print(f'\nCongratulations {name}! You won the match.')
    elif p_score == y_score :
        print(f'\n{name}! The match has been draw between you and program.')
    else :
        print(f'Sorry {name}! You couldn\'t pass this game.You can play it again.')
        
except :
    print('\nEnter valid input.')
```
